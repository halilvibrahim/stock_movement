```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

df = pd.read_csv('googl.csv')  
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
plt.figure(figsize=(10, 6))

plt.plot(df.index, df['Ensemble_Mamba'], color='blue', label='Ensemble Mamba')
plt.plot(df.index, df['Mamba'], color='green', label='Mamba')
plt.plot(df.index, df['BH_Value'], color='red', label='Buy & Hold')

plt.xlabel('Day')
plt.ylabel('Capital ($)')
plt.title('')


plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=15))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.xticks(rotation=45)
plt.xlim([df.index.min(), df.index.max()])
plt.legend()
plt.tight_layout()
plt.savefig('profit_googl.png', format='png', dpi=300)
plt.show()
```
